```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('/content/movies.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
features=['genres','keywords','tagline', 'spoken_languages','cast', 'director']
```

```python
data.isnull().sum()
```

```python
for feature in features:
  data[feature]=data[feature].fillna('')
```

```python
#combining all features into one
combined_feature=data['genres']+''+data['keywords']+''+data['tagline']+''+data['spoken_languages']+''+data['cast']+''+data[ 'director']
```

```python
vectorize=TfidfVectorizer()
```

```python
vec_feature=vectorize.fit_transform(combined_feature)
```

```python
print(vec_feature)
```

```python
similarity=cosine_similarity(vec_feature)
```

```python
similarity.shape
```

```python
title_list=data['title'].tolist()
```

```python
title_input=input('Enter the title: ')
```

```python
find_match=difflib.get_close_matches(title_input,title_list)
```

```python
find_match
```

```python
close_match=find_match[0]
close_match
```

```python
#getting index
index_val=data[data.title==close_match]['index'].values[0]
print(index_val)
```

```python
data[data['index']==index_val]
```

```python
similarity_li=list(enumerate(similarity[index_val]))
print(similarity_li)
```

```python
sorted_list=sorted(similarity_li,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_list)
```

```python
i=1
for movie in sorted_list:
  index=movie[0]
  title_index=data[data.index==index]['title'].values[0]
  if i<=20:
    print(i,'.',title_index)
    i+=1
  else:
    break
```

Final code

```python
title_input=input('Enter the title: ')
find_match=difflib.get_close_matches(title_input,title_list)
close_match=find_match[0]
index_val=data[data.title==close_match]['index'].values[0]
similarity_li=list(enumerate(similarity[index_val]))
sorted_list=sorted(similarity_li,key=lambda x:x[1],reverse=True)
i=1
for movie in sorted_list:
  index=movie[0]
  title_index=data[data.index==index]['title'].values[0]
  if i<=20:
    print(i,'.',title_index)
    i+=1
  else:
    break
```
